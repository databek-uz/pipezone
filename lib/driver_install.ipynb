```python
import urllib.request
import shutil
```

```python
def install_driver_dynamic(jar_url, jar_name):
    local_path = f"/databricks/jars/{jar_name}"

    print(f"Downloading {jar_url} -> {local_path}")
    with urllib.request.urlopen(jar_url) as response, open(local_path, 'wb') as out_file:
        shutil.copyfileobj(response, out_file)

    print("Driver installed:", local_path)
    return local_path
```

```python
install_driver_dynamic(
    jar_url="https://repo1.maven.org/maven2/com/mysql/mysql-connector-j/9.5.0/mysql-connector-j-9.5.0.jar",
    jar_name="mysql-driver.jar"
)
```
